Posibles objetivos de este EDA:
* ¿Cómo ha evolucionado la inflación a lo largo del tiempo, y qué patrones subyacentes (tendencias, estacionalidades) podemos identificar que podrían influir en la política monetaria futura?

* Preparar la data para hacer análisis predictivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar datos de inflación
url = 'https://fred.stlouisfed.org/graph/fredgraph.csv?bgcolor=%23e1e9f0&chart_type=line&drp=0&fo=open%20sans&graph_bgcolor=%23ffffff&height=450&mode=fred&recession_bars=on&txtcolor=%23444444&ts=12&tts=12&width=1320&nt=0&thu=0&trc=0&show_legend=yes&show_axis_titles=yes&show_tooltip=yes&id=CPIAUCSL&scale=left&cosd=1947-01-01&coed=2024-07-01&line_color=%234572a7&link_values=false&line_style=solid&mark_type=none&mw=3&lw=2&ost=-99999&oet=99999&mma=0&fml=a&fq=Monthly&fam=avg&fgst=lin&fgsnd=2020-02-01&line_index=1&transformation=lin&vintage_date=2024-08-29&revision_date=2024-08-29&nd=1947-01-01'
inflation_data = pd.read_csv(url, parse_dates=['DATE'], index_col='DATE')
```

```python
# Información general sobre el DataFrame (ustedes)
print("Información del DataFrame:")
print(inflation_data.info())
```

```python
# Ver las primeras filas (ustedes)
print("\nPrimeras filas del DataFrame:")
print(inflation_data.head())
```

```python
# Ver las últimas filas (ustedes)
print("\nÚltimas filas del DataFrame:")
print(inflation_data.tail())
```

```python
# Verificar valores faltantes (ustedes)
missing_values = inflation_data.isnull().sum()
print("\nValores faltantes por columna:")
print(missing_values)
```

```python
# Descriptivas (ustedes)
print(inflation_data.describe())
```

```python
# Gráfica de la serie temporal (ustedes)
plt.figure(figsize=(10, 6))
plt.plot(inflation_data, label='Inflación (Índice de Precios al Consumidor)')
plt.title('Serie Temporal de Inflación')
plt.xlabel('Fecha')
plt.ylabel('Índice de Precios')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie temporal
result = seasonal_decompose(inflation_data, model='multiplicative')

# Visualización de la descomposición
result.plot()
plt.show()
```

1. Tendencia (Trend):

La serie temporal muestra una clara tendencia creciente a lo largo del tiempo. Esto indica que, en general, los valores de la serie han estado aumentando de manera sostenida desde 1950 hasta 2020. Esta tendencia creciente podría estar relacionada con un fenómeno económico subyacente, como el crecimiento de precios o índices económicos.
2. Estacionalidad (Seasonal):

El componente estacional muestra patrones repetitivos en intervalos regulares, lo cual es típico de datos que presentan estacionalidad. Estos patrones parecen ser muy consistentes a lo largo del tiempo, lo que indica que hay ciclos regulares que afectan la serie. Estos ciclos podrían representar fluctuaciones anuales o trimestrales comunes en datos económicos, como la demanda estacional de ciertos productos o servicios.
3. Residuos (Resid):

Los residuos (o componente aleatorio) parecen ser bastante pequeños y no muestran un patrón evidente, lo que es deseable. Esto sugiere que la mayor parte de la variación en la serie temporal se explica por la tendencia y la estacionalidad, y que no queda mucho ruido o variabilidad no explicada.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie temporal
result = seasonal_decompose(inflation_data, model='additive')

# Visualización de la descomposición
result.plot()
plt.show()
```

Los residuos nos indican lo que el modelo (tendencia y estacionalidad) no nos puede explicar. Lo cual quiere decir que algo más pasó, y afectó a los precios. Acá podemos ver que algo claramente pasó en 2008-2009 y en 2020-2021.

```python
from pandas.plotting import lag_plot
from statsmodels.graphics.tsaplots import plot_acf

# Gráfico de retrasos (Lag plot)
plt.figure(figsize=(6, 6))
lag_plot(inflation_data)
plt.title('Lag Plot de la Inflación')
plt.show()

```

Este Lag Plot indica una alta autocorrelación en la serie temporal de inflación, sugiriendo que la serie tiene un comportamiento muy predecible y posiblemente una tendencia subyacente significativa. Este tipo de patrón es típico en series temporales económicas como la inflación, donde los valores de un período están estrechamente relacionados con los valores del período anterior.

IPC(t), IPC(t-1), IPC(t-2), IPC(t-3)....

```python
# Gráfico de autocorrelación
plt.figure(figsize=(10, 6))
plot_acf(inflation_data, lags=24)
plt.title('Función de Autocorrelación de la Inflación')
plt.show()
```

El gráfico ACF muestra una fuerte autocorrelación en todos los retardos, lo que indica que la serie de inflación tiene una tendencia pronunciada y probablemente no es estacionaria. Esto implica que se necesitarían transformaciones adicionales para preparar la serie para un análisis y modelado más riguroso, como la diferenciación para eliminar la tendencia y hacer que la serie sea más estacionaria.

```python
from statsmodels.tsa.stattools import adfuller

# Prueba de Dickey-Fuller aumentada para verificar la estacionariedad
result = adfuller(inflation_data.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Interpretación: Si el p-value es menor que 0.05, la serie es estacionaria.
```

```python
import numpy as np
```

```python
# Aplicar una transformación logarítmica para estabilizar la varianza
inflation_log = np.log(inflation_data)

# Aplicar diferenciación para remover la tendencia
inflation_diff = inflation_log.diff().dropna()

# IPC(t) => log(IPC(t)) - log(IPC(t-1))
# IPC(dic-23) => log(IPC(dic-23)) - log(IPC(nov-23))

# Visualizar la serie diferenciada
plt.figure(figsize=(10, 6))
plt.plot(inflation_diff, label='IPC Diferenciada')
plt.title('Serie Temporal de IPC (Diferenciada)')
plt.xlabel('Fecha')
plt.ylabel('Diferencia Logarítmica')
plt.legend()
plt.show()
```

El gráfico muestra que la serie temporal de inflación, después de la diferenciación y la transformación logarítmica, es mucho más estable y no muestra una tendencia obvia, lo que sugiere que la serie puede ser estacionaria. Esto es crucial para el modelado predictivo, ya que permite aplicar modelos como ARIMA para hacer predicciones basadas en la serie transformada. Además, las fluctuaciones en ciertos períodos históricos pueden indicar eventos económicos importantes que valdría la pena investigar más a fondo.

Acá podríamos estimar un ARIMA y hacer una predicción de cómo va a evolucionar la inflación. No es el cometido de este curso.

```python
from statsmodels.tsa.stattools import adfuller

# Prueba de Dickey-Fuller aumentada para verificar la estacionariedad
result = adfuller(inflation_diff.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from pandas.plotting import lag_plot
from statsmodels.graphics.tsaplots import plot_acf

# Gráfico de retrasos (Lag plot)
plt.figure(figsize=(6, 6))
lag_plot(inflation_diff)
plt.title('Lag Plot de la Inflación')
plt.show()

```

```python
# Descomposición de la serie temporal
result = seasonal_decompose(inflation_diff, model='additive')

# Visualización de la descomposición
result.plot()
plt.show()
```
