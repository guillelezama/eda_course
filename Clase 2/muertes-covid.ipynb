```python
# Una forma de visualizarlo! Queremos que para las fechas que hay un valor positivo, el color sea uno.
# Para las fechas que el valor sea 0, el color sea otro.

import seaborn as sns
import matplotlib.pyplot as plt

df['year'] = pd.to_datetime(df['date']).dt.year
#Binarizamos el dataset: 1 cuando es 0, 0 cuando es positiva
highlight_df = (df[df['year']==2021]['Uruguay'] == 0).astype(int).to_frame()

# Queremos tener fechas en el eje de las y.
highlight_df['date'] = df[df['year']==2021]['date']

# La fecha como indice.
highlight_df.set_index('date', inplace=True)

# Creamos el heatmap que tiene como input una tabla con un indice, y la variable Uruguay.
plt.figure(figsize=(10, 6))
ax = sns.heatmap(highlight_df, cmap='viridis', yticklabels=True)

# Los y ticks cada 90 dias (para que no quede sobrepoblado)
y_ticks = range(0, len(highlight_df), 90)
ax.set_yticks(y_ticks)
ax.set_yticklabels(highlight_df.index[y_ticks])

plt.title('Fechas con Uruguay = 0')
plt.ylabel('Date')
plt.show()
```

## Analisis Grafico

Para complementar las estadísticas descriptivas, podemos utilizar gráficas.

Presentar:
1. Distribución de Nuevas Muertes por País (Histograma)
2. Evolución Temporal de Muertes en Uruguay y Argentina (Serie Temporal)
3. Boxplot para Comparar la Distribución de Nuevas Muertes por Año
4. Visualización de Correlación. Explorar si hay correlación entre las muertes en Uruguay y Argentina, o cómo se relacionan las muertes en diferentes países.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))

# Histograma para Uruguay
plt.subplot(1, 2, 1)
sns.histplot(df['Uruguay'], kde=True, color='blue')
plt.title('Distribución de Nuevas Muertes Diarias en Uruguay')
plt.xlabel('Nuevas Muertes Diarias')
plt.ylabel('Frecuencia')

# Histograma para Argentina
plt.subplot(1, 2, 2)
sns.histplot(df['Argentina'], kde=True, color='orange')
plt.title('Distribución de Nuevas Muertes Diarias en Argentina')
plt.xlabel('Nuevas Muertes Diarias')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='date', y='Argentina', data=df, label='Argentina', color='orange')
sns.lineplot(x='date', y='Uruguay', data=df, label='Uruguay', color='blue')

plt.title('Evolución Temporal de Nuevas Muertes Diarias por COVID-19')
plt.xlabel('Fecha')
plt.ylabel('Nuevas Muertes Diarias')

# Reducir el número de x-ticks
plt.xticks(pd.date_range(start=df['date'].min(), end=df['date'].max(), freq='6M').strftime('%Y-%m'), rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='Uruguay', data=df, color='lightblue')
plt.title('Distribución de Nuevas Muertes por Año en Uruguay')
plt.xlabel('Año')
plt.ylabel('Nuevas Muertes Diarias')
plt.show()
```

```python
plt.figure(figsize=(10, 8))

# Scatter plot con mapa de densidad (hexbin)
plt.hexbin(df['Uruguay'], df['Argentina'], gridsize=50, cmap='Blues', reduce_C_function=np.mean)
plt.colorbar(label='Densidad')
plt.title('Relación entre Nuevas Muertes Diarias en Uruguay y Argentina')
plt.xlabel('Nuevas Muertes en Uruguay')
plt.ylabel('Nuevas Muertes en Argentina')

plt.show()
```

```python
# Filtrar días donde ambos países tienen 0 muertes, ya que hay mucho overlap en 0
df_filtered = df[(df['Uruguay'] > 0) | (df['Argentina'] > 0)]

plt.figure(figsize=(10, 8))

# Scatter plot con hexbin para días con al menos una muerte en Uruguay o Argentina
plt.hexbin(df_filtered['Uruguay'], df_filtered['Argentina'], gridsize=50, cmap='Blues', reduce_C_function=np.mean)
plt.colorbar(label='Densidad')
plt.title('Relación entre Nuevas Muertes Diarias en Uruguay y Argentina (sin ceros)')
plt.xlabel('Nuevas Muertes en Uruguay')
plt.ylabel('Nuevas Muertes en Argentina')

plt.show()
```

```python
# Se pueden agregar datos por semana
df['week'] = pd.to_datetime(df['date']).dt.to_period('W').apply(lambda r: r.start_time)
df_weekly = df.groupby('week').sum().reset_index()

plt.figure(figsize=(10, 8))

sns.scatterplot(x='Uruguay', y='Argentina', data=df_weekly, alpha=0.6, s=100)
plt.title('Relación entre Nuevas Muertes Semanales en Uruguay y Argentina')
plt.xlabel('Nuevas Muertes Semanales en Uruguay')
plt.ylabel('Nuevas Muertes Semanales en Argentina')

plt.show()
```
